```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,5,10)
print(x)
ax.plot(x,np. sin(x))
ax.plot(x,np. cos(x))
ax.plot(x,np. tan(x))
ax.set_title('basii')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,5,10)
print(x)
ax.plot(x,np. sin(x),label='sin', color='red')
ax.plot(x,np. cos(x),label='cos', color='blue')
ax.plot(x,np. tan(x),label='tan', color='pink')
ax.set_title('multiple titles')
ax.set_xlabel('range')
ax.set_ylabel('amplitude')
```
